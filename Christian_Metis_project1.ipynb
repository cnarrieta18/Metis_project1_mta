```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 20)
pd.set_option('display.precision', 3)
```

```python
url = 'http://web.mta.info/developers/data/nyct/turnstile/turnstile_190629.txt'
df = pd.read_csv(url, header = 0, sep=',')
df.columns = map(str.lower, df.columns) # - make headers lowercase
```

```python
df.head(5)
```

```python
df['time'] = pd.to_datetime(df['time'])
df.head(5)
```

```python
#clean column headers
df.columns = df.columns.str.strip()
df.columns
```

```python
#check trend of entries per station
y_entries = df[df.station == 'RIT-ROOSEVELT'].entries  #change station to check different station trends
x_entries = np.arange(0,len(y_entries),1)
plt.plot(x_entries,y_entries)
```

```python
#David's code
result_1 = df.groupby(['c/a','unit','scp','station'])['entries'].apply(lambda x: x.max() - x.min()).reset_index()
result_2 = result_1.groupby('station')['entries'].sum()
result_2
```
